```python
# Dzień 5 - Filtrowanie danych

import pandas as pd
df = pd.read_csv(
  r'C:\Users\damia\Desktop\coderslab - Data scientist\3. Pandas\Zjazd_3_-_pliki_do_zadań\Zjazd 3 - pliki do zadań\01_Dane\ceny_produktow_renamed.csv',
  sep=';',
  encoding='UTF-8',
  decimal='.'
)
df.head()
```

```python
# 1. W kolumnie date pojawiły się dane z roku 1888 - '1888-0'

df = df.query("date == '1888-0'")
df.head(17)
```

```python
# 2. Sprawdź, czy nie pojawiły się podobne błędy w kolumnie date, jednak z wartościami z przyszłości

df = df.query("date > '2023-0'")
df.head()
```

```python
# 3. W kolumnie value pojawiła się zbyt wysoka wartość – znajdź ją i zlokalizuj wiersz, w którym jest wpisana (używając query)

df = df.query("value == 3000")
df.head()
```
