```python
# Dzień 5 Pandas - Obróbka danych - Część 1 - Najwyższa i najniższa cena

import pandas as pd
df = pd.read_csv(
  r'C:\Users\damia\Desktop\coderslab - Data scientist\3. Pandas\Zjazd_3_-_pliki_do_zadań\Zjazd 3 - pliki do zadań\01_Dane\ceny_produktow_renamed.csv',
  sep=';',
  encoding='UTF-8',
  decimal='.'
)
df.head()
```

```python
# wyrzucenie wartosci 0 i 3000
df_new = df.loc[(df['value'] != 0) & (df['value'] != 3000) & (df['province'] == 'POLSKA') & (df['currency'] == 'zł')]
```

```python
# najwyzsza cena
df_new.sort_values(by=['value'], ascending=[False]).iloc[0]
```

```python
# najnizsza cena
df_new.sort_values(by=['value'], ascending=[True]).iloc[0]
```

```python
# Posortuj dane według następującego klucza:

#     według kolumny product_types – malejąco
#     według kolumny value – rosnąco

df_new.sort_values(by=['product_types', 'value'], ascending=[False, True]).head(20)
```
